```python

import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#import mpld3
#mpld3.enable_notebook()
```

```python
Ford = pd.read_csv('Ford.csv', parse_dates = True, index_col = 'Date')
Honda = pd.read_csv('HONDA.csv', parse_dates = True, index_col = 'Date')
Tesla = pd.read_csv('TSLA.csv', parse_dates = True, index_col = 'Date')
```

```python
Ford.insert(0, 'Symbol', 'Ford')
Honda.insert(0, 'Symbol', 'Honda')
Tesla.insert(0, 'Symbol', 'TSLA')
```

```python
data = pd.concat([Ford, Honda,Tesla]).sort_index()        #combined 3 companies
data
```

```python
#Simple moving average
```

```python
data['SMA20'] = data['Adj Close'].rolling(20).mean()
data['SMA120'] = data['Adj Close'].rolling(120).mean()
data
```

```python
data = data.drop(columns = ['High', 'Low', 'Open', 'Close'])   # remove the unneeded columns.
```

```python
data
```

```python
data['price_yesterday'] = data['Adj Close'].shift(3)
data.head(30)
```

```python
data['change'] = data['Adj Close'] / data['price_yesterday']
data.head()
```

```python
data['invested_SMA'] = np.where(data['SMA20'] > data['SMA120'], 1, 0)
data
```

```python
sma=data[data['invested_SMA'] == 1]# Will produce only that part of the data frame where invested_SMA is equal to 1.
sma
```

```python
#BUY AND HOLD
```

```python
data['buy_and_hold'] = np.cumprod(data['change'])
data
```

```python
sma['return'] = np.cumprod(sma['change'])
sma
```

```python
#############Exponential AVerage
```

```python
data['EWM12'] = data['Adj Close'].ewm(span =12, adjust = False).mean()
data['EWM26'] = data['Adj Close'].ewm(span =26, adjust = False).mean()
```

```python
data['invested_EWM'] = np.where(data['EWM12'] > data['EWM26'], 1, 0)
data
```

```python
ewm = data[data['invested_EWM'] == 1]
ewm
```

```python
ewm['return'] = np.cumprod(ewm['change'])
ewm
```
